```python
import pandas as pd
```

```python
consolidado=pd.read_excel("consolidado_final_vih.xlsx", index=False)
consolidado
```

```python
gestantes=consolidado[consolidado["EDAD"] == "Gestantes"]
gestantes.to_excel("consolidado_gestantes.xlsx", index=False)
gestantes
```

```python
trans=consolidado[consolidado["EDAD"] == "Trans"]
trans.to_excel("consolidado_trans.xlsx", index=False)
trans
```

```python
poblacion=consolidado[consolidado["EDAD"] == "Población Inmigrante"]
poblacion.to_excel("consolidado_poblacion_inmigrantes.xlsx", index=False)
poblacion
```

```python
pueblos=consolidado[consolidado["EDAD"] == "Pueblos Originarios"]
pueblos.to_excel("consolidado_pueblos_originarios.xlsx", index=False)
pueblos
```

```python
edades1 = consolidado[consolidado["EDAD"] != "Gestantes"]
edades2 = edades1[edades1["EDAD"] != "Trans"]
edades3 = edades2[edades2["EDAD"] != "Población Inmigrante"]
edades4 = edades3[edades3["EDAD"] != "Pueblos Originarios"]
edades4.to_excel("consolidado_edades.xlsx", index=False)
```
